# Local Capsule Example Using Django

To run this example

- install redis by running "brew install redis"
- start redis by running "redis-server"
- install dependencies by running "pip install -r requirements.txt"
- start local capsule server by running "sh build_and_run_django.sh"

```python
from capsule.django_client import LocalDjangoCapsuleClient
import syft as sy
import requests
```

```python
r = requests.get('http://localhost:9000/keygen/45/paillier/')
r.content.decode()
```

```python
cc = LocalDjangoCapsuleClient()
pk = cc.keygen()
```

```python
x = pk.ones(10)
y = sy.ones(10)*0.2
```

```python
out = x.dot(y)
```

```python
x.decrypt(cc)
```

```python
y.decrypt(cc)
```

```python
out.decrypt(cc)
```

```python
out = cc.bootstrap(x)
```
